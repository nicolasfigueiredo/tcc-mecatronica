```python
with open("sobrenomes.txt") as f:
    sobrenomes = [line[:-1] for line in f]
```

```python
sobrenomes
```

```python
nomes_completos = []
for i in range(len(sobrenomes)-1):
    nomes_completos.append(nomes[i] + ' ' + sobrenomes[i])
    nomes_completos.append(nomes[i] + ' ' + sobrenomes[i+1])
```

```python
nomes_completos
```

```python
nomes[74]
```

```python
for i in range(74, len(nomes)):
    nomes_completos.append(nomes[i] + ' ' + sobrenomes[i-70])
    nomes_completos.append(nomes[i] + ' ' + sobrenomes[i-69])
```

```python
nomes_completos
```

```python
with open('nomes_completos.txt', 'w') as f:
    for nome in nomes_completos:
        f.write(nome + '\n')
```
